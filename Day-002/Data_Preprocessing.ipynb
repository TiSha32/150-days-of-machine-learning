# Data Leak prevention


```python
data = pd.read_csv('/content/drive/MyDrive/Housing.csv')
data.head()
```

```python
pd.DataFrame({'Missing_values' : data.isnull().sum(),
             'Data_Types' : data.dtypes})
```

```python
#dependency split
x = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
#train test split
x_train,x_test, y_train,y_test = train_test_split(x,y, random_state = 42, test_size = 0.3)
```

```python
# x_train sample
x_train.head()
#print(f'Shape of x_train: {x_train.shape}, Shape of x_test: {x_test.shape}')
```

```python
#train test encoding


#train encoding
encoder = OneHotEncoder(sparse_output = False, dtype = 'i2', drop = 'first')
cat_cols = [cols for cols in x_train.columns if x_train[cols].dtype == 'O']
x_train_cat_encoded = encoder.fit_transform(x_train[cat_cols])
x_train_cat_df = pd.DataFrame(x_train_cat_encoded, index = x_train.index, columns = encoder.get_feature_names_out())
x_train_encoded = pd.concat([x_train_cat_df, x_train.drop(cat_cols, axis = 1)], axis = 1)
x_train_encoded.head()


#test encoding

x_test_cat_encoded = encoder.transform(x_test[cat_cols])
x_test_cat_df = pd.DataFrame(x_test_cat_encoded, index = x_test.index, columns = encoder.get_feature_names_out())
x_test_encoded = pd.concat([x_test_cat_df, x_test.drop(cat_cols, axis = 1)], axis = 1)
x_test_encoded.head()
```

```python
# train test scaling

scaled = StandardScaler()
x_train_scaled = scaled.fit_transform(x_train_encoded)
x_train_scaled_df = pd.DataFrame(x_train_scaled, index = x_train_encoded.index, columns = scaled.get_feature_names_out())
x_train_scaled_df.head()

x_test_scaled = scaled.fit_transform(x_test_encoded)
x_test_scaled_df = pd.DataFrame(x_test_scaled, index = x_test_encoded.index, columns = scaled.get_feature_names_out())
round(x_test_scaled_df.describe(),2)
```
